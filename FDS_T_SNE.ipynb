```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data  # Features (4 features for each flower)
y = iris.target  # Actual labels (for visualization purposes)

# Step 2: Standardize the data (important for K-means)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Perform K-means clustering (using k=3 as there are 3 classes in the Iris dataset)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
y_kmeans = kmeans.predict(X_scaled)

# Step 4: Reduce dimensionality using t-SNE (4D -> 2D)
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

# Step 5: Visualize the result
plt.figure(figsize=(8, 6))

# Scatter plot using t-SNE for 2D visualization
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y_kmeans, cmap='viridis', label="K-means Cluster")
plt.colorbar(label='Cluster label')
plt.title('t-SNE visualization of K-means clustering on Iris dataset')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')

plt.show()
```
