```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from google.colab import drive
drive.mount('/content/drive')
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Salary']=df['Salary'].fillna(df['Salary'].mean())
```

```python
df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['Age'] = scaler.fit_transform(df[['Age']])
df['Salary'] = scaler.fit_transform(df[['Salary']])
print(df)
```

```python
x=np.array(df['Age'])
y=np.array(df['Salary'])
mean_x=np.mean(x) #mean
mean_y=np.mean(y)
```

```python
covarience=np.sum((x-mean_x)*(y-mean_y))/len(x)
print(covarience)
```

```python
std_x=np.std(x)
std_y=np.std(y)
print(std_x)
print(std_y)
```

```python
perason_r=covarience/(std_x*std_y)
print(perason_r)
```

```python
x_points =np.random.randint(0,10,10)
y_line= 3*x_points+19
plt.scatter(x_points,y_line)
plt.show()
person_r1=np.corrcoef(x_points,y_line)
print(person_r1)
```

```python
x_points =np.random.randint(0,10,10)
y_line= -3*x_points
plt.scatter(x_points,y_line)
plt.show()
person_r2=np.corrcoef(x_points,y_line)
print(person_r2)
```

```python
x = np.random.rand(1, 11)
y_positive = 3 * x + 19
y_negative = -3 * x

plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.plot(x, y_positive, 'bo-', label='y = 3x + 19')
plt.title('y = 3x+19')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
#plt.legend()


plt.subplot(1, 2, 2)
plt.plot(x, y_negative, 'ro-', label='y = -3x')
plt.title('y = -3x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
#plt.legend()
plt.tight_layout()
plt.show()
```
