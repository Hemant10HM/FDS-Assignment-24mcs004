```python
#Design a gradiant descent algorithm
#get random pints whose equation is 3X+0.7
import numpy as np

# Generate 30 random x values
x_values = np.random.uniform(-10, 10, 30)  # You can change the range as needed

# Calculate the corresponding y values using the equation y = 3x + 0.7
y_values = 3 * x_values + 0.7

# Store the data in a list of tuples
data = list(zip(x_values, y_values))

# Optionally, print the stored data
for x, y in data:
    print(f"x: {x:.2f}, y: {y:.2f}")
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_values,y_values)
plt.show()
w_old=np.random.uniform(-10,10)
b_old=np.random.uniform(-10,10)
print(w_old,b_old)
y_cap=w_old*x_values+b_old
Loss = 1/2 * np.mean((y_cap - y_values)**2)
for i in range(500):
    w_new=w_old-0.01*np.mean((y_cap-y_values)*x_values)
    b_new=b_old-0.01*np.mean((y_cap-y_values))
    y_cap=w_new*x_values+b_new
    Loss = 1/2 * np.mean((y_cap - y_values)**2)
    w_old=w_new
    b_old=b_new
    print(Loss)
print(w_new,b_new)
# Plotting the results
plt.scatter(x_values, y_values, color='blue', label='Data Points')
plt.plot(x_values, y_cap, color='orange', label='Fitted Line', linewidth=2)
plt.title('Gradient Descent Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
```
